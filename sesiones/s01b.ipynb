```python
import pandas

articulos = pandas.read_csv("../datasets/articles.csv")

articulos
```

```python
corpus = "\n\n".join(articulos["article"].map(str).values)

with open("../datasets/corpus_nytimes.txt", "w") as file:
    file.write(corpus)
```

```python
from tokenizers import BertWordPieceTokenizer

tokenizer = BertWordPieceTokenizer()

tokenizer.train(files=["../datasets/corpus_nytimes.txt"], vocab_size=30_000, min_frequency=1)

tokenizer.save_model(".", "../models/nytimes_bpe_tokenizer")
```

```python
# Test the tokenizer
encoded = tokenizer.encode("somethings")
print("Encoded:", encoded.tokens)

# Decode tokens back to text
decoded = tokenizer.decode(encoded.ids)
print("Decoded:", decoded)
```
