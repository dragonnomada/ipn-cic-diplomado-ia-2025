```python
from diffusers import StableDiffusionPipeline
import torch
```

```python
pipeline = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")
# runwayml/stable-diffusion-v1-5
# stabilityai/stable-diffusion-2
# stabilityai/stable-diffusion-2-1
# stabilityai/stable-diffusion-2-base
# stabilityai/stable-diffusion-xl-base-1.0
# stabilityai/stable-diffusion-xl-refiner-1.0
# stabilityai/sdxl-turbo
# stabilityai/stable-diffusion-3-medium-diffusers
# stabilityai/stable-diffusion-3-large-diffusers
# stabilityai/stable-diffusion-3.5-large

pipeline.to("cpu")
```

```python
prompt = "A Pikachu Charmander pokemon inside a pokeball"
```

```python
result = pipeline(prompt)
```

```python
image = result.images[0]

image.show()
```

```python
image.save("pikachu.png")
```

```python
from PIL import Image
import numpy
from matplotlib.pyplot import imshow
```

```python
image_input = Image.open("ameli_angstrom.jpeg")
image_input.save("ameli_angstrom.png")
image_input = Image.open("ameli_angstrom.png").convert("RGB")

imshow(image_input)
```

```python
type(image_input)
```

```python
mask = numpy.zeros((image_input.height, image_input.width))

mask[0:300, 0:300] = 255

imshow((mask + image_input.convert("L")) / 2, cmap="inferno")
```

```python
from diffusers import StableDiffusionInpaintPipeline
```

```python
pipeline_inpaint = StableDiffusionInpaintPipeline.from_pretrained("stable-diffusion-v1-5/stable-diffusion-inpainting")

pipeline_inpaint.to("cuda")
```

```python
prompt = "Fill the region with a flower or a tree"
```

```python
result2 = pipeline_inpaint(prompt=prompt, image=image_input, mask_image=mask)
```

```python
image2 = result2.images[0]

imshow(image2)
```

```python
image2.save("ameli_angstrom_fix1.png")
```

```python
prompt = "Fill missing parts with a pikachu"

image3 = pipeline_inpaint(prompt=prompt, image=image_input, mask_image=mask).images[0]

image3.save("ameli_angstrom_fix2.png")
```

```python
prompt = "Fill missing parts with a pikachu"

image4 = pipeline_inpaint(prompt=prompt, image=image_input, mask_image=(255 - mask)).images[0]

image4.save("ameli_angstrom_fix3.png")
```

```python
mask = numpy.zeros((image_input.height, image_input.width))

mask[:, 0:400] = 255

imshow((mask + image_input.convert("L")) / 2, cmap="inferno")
```

```python
prompt = "Create a picachu on the top"

image5 = pipeline_inpaint(prompt=prompt, image=image_input, mask_image=(255 - mask)).images[0]

image5.save("ameli_angstrom_fix4.png")
```

```python
prompt = "Create a picachu on the top"

image = pipeline_inpaint(prompt=prompt, image=image_input, mask_image=mask).images[0]

image.save("ameli_angstrom_fix5.png")
```
