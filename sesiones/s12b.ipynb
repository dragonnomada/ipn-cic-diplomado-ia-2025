```python
import google.generativeai as generativeai

with open("generativeai.key", "r") as file:
    apiKey = file.read()

generativeai.configure(api_key=apiKey)

generativeai
```

```python
client = generativeai.GenerativeModel(model_name="models/gemini-2.5-flash")

client
```

```python
response = client.generate_content("Hola, puedes explicarme como usar generativeai.GenerativeModel para imágenes de nano banana")

response
```

```python
print(response.candidates[0].content.parts[0].text)
```

```python
for model in generativeai.list_models():
    print(model.name)
```

## Generación de imágenes

```python
client = generativeai.GenerativeModel(model_name="models/gemini-2.5-flash-image-preview")
client
```

```python
response = client.generate_content("Genera una imagen que fusione a pikachú y charmander en un pokebola")
response
```

```python
for part in response.candidates[0].content.parts:
    if "inline_data" in part:
        # print(part)
        image = part.inline_data.data
        with open("pikacharmander_gemini.png", "wb") as file:
            file.write(part.inline_data.data)
```
